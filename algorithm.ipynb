```python
import matplotlib.pyplot as plt
# Years and profits
years = [2018, 2019, 2020, 2021, 2022, 2023]
profits = [20000, 25000, 18000, 30000, 35000, 40000]
# Plotting line graph
plt.plot(years, profits, marker='o')
# Labels and title
plt.xlabel("Years")
plt.ylabel("Profit")
plt.title("Company Profits Over Years")
# Show graph
plt.show()
```
