```python
import matplotlib.pyplot as plt

# Employee IDs
emp_ids = [101, 102, 103, 104, 105]

# Salaries of two departments
dept1_salaries = [40000, 45000, 50000, 55000, 60000]
dept2_salaries = [38000, 47000, 52000, 53000, 58000]

# Set width for bars
bar_width = 0.35

# X-axis positions
import numpy as np
x = np.arange(len(emp_ids))

# Plotting the bars
#plt.bar(x_positions, heights, width=bar_width, label='name')
             #x_positions → The horizontal positions of the bars (along the x-axis).
             # heights → The values you want to show (the length of each bar, along the y-axis).
             # width → The width of each bar.
            #label → Name for the legend

plt.bar(x - bar_width/2, dept1_salaries, width=bar_width, label='Dept 1')
plt.bar(x + bar_width/2, dept2_salaries, width=bar_width, label='Dept 2')

# Labels and title
plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.title("Salary Comparison of Two Departments")
plt.xticks(x, emp_ids)
plt.legend()

# Show graph
plt.show()


```
